```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
# generate different random samples
nsamp = [10, 50, 100, 1000]
mu = 0.5
sigma = 1

# plot subfigures
plt.figure(figsize=(10,8))

for i in range(len(nsamp)):
    plt.subplot(2,2,i+1)
    x = st.norm.rvs(loc=mu, scale=sigma, size=nsamp[i])
    plt.hist(x)
    plt.title("N samples={:.0f}".format(nsamp[i]))
```

## Example of p-value bullshit

```python
nsamp = int(1e4)
mu = 5
sigma = 1


np.random.seed(42)
x = np.zeros(nsamp)
x[:-1] = st.norm.rvs(loc=mu, scale=sigma, size=(nsamp-1))
x[-1] = 100 # large outlier, Fat Tail

# no outlier
xNormal = x[:-1].copy()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(xNormal, bins=30)
plt.title("excluding x=100")

plt.subplot(1,2,2)
plt.hist(x, bins=30)
plt.title("full data for x")

plt.show()
```
